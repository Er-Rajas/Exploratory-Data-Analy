```python
plt.hist(df["AGE"],bins=30, color = 'skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# First check unique values
# Convert ICU values to numeric, excluding 97 and 99
df["ICU_1"] = pd.to_numeric(df["ICU"], errors='coerce')
df["ICU_1"] = df["ICU_1"].replace([97, 99], np.nan)

# Create plot
sns.countplot(data=df.dropna(), x='ICU_1', palette='Set2')
plt.title('ICU Admission Count')
plt.xlabel('ICU (1 = Admitted, 0 = Not Admitted)')
plt.ylabel('Count')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
correlation_matrix =numeric_df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm', fmt='.2f')
plt.title('Correaltion Matrix')
plt.show()
```

```python
df['DIED'] = df['DATE_DIED'] != '9999-99-99'

sns.countplot(data=df, x='DIED', palette ='cool')
plt.title('Survival vs Death Count')
plt.xlabel("Died (True/False)")
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='SEX',data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Sex (1 = Male, 2 = Female)')
plt.ylabel('Count')
plt.show()
```

```python
# Check data
df["ICU_1"] = pd.to_numeric(df["ICU"], errors='coerce')
df["ICU_1"] = df["ICU_1"].replace([97, 99], np.nan)

plt.figure(figsize=(8, 6))
sns.boxplot(x='ICU_1', y='AGE', data=df, palette='Set2')
plt.title('ICU Admission vs Age')
plt.xlabel('ICU (1=Admitted, 2=Not Admitted)')
plt.ylabel('Age')
plt.show()
```

```python
comorbidities = ['DIABETES', 'COPD', 'ASTHMA', 'HIPERTENSION', 'CARDIOVASCULAR', 'OBESITY']
death_rates = df[df['DIED'] == True][comorbidities].sum()

plt.figure(figsize=(10, 6))
death_rates.sort_values().plot(kind='barh', color='salmon')
plt.title('Deaths by Comorbidities')
plt.xlabel('Number of Deaths')
plt.ylabel('Comorbidities')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(df[df['DIED'] == True]['AGE'], fill=True, color='red', label='Died')
sns.kdeplot(df[df['DIED'] == False]['AGE'], fill=True, color='blue', label='Survived')
plt.title('Age Distribution of Deceased vs Survivors')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
